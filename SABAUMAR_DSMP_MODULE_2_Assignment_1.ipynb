## **SECTION 1: SQL Basics**

 **SELECT**



```
SELECT * FROM employees;

```
 **WHERE**



```
SELECT * FROM employees WHERE salary > 50000;
```

 **ORDER BY**



```
SELECT * FROM employees ORDER BY name ASC;

```





## **SECTION 2: Intermediate Concepts**

 **INSERT INTO**



```
INSERT INTO employees (name, department, salary) VALUES ('Ayesha', 'HR', 45000);

```

 **UPDATE**



```
UPDATE employees SET salary = 50000 WHERE name = 'Ayesha';

```

 **DELETE**



```
DELETE FROM employees WHERE name = 'Ayesha';

```

 **COUNT, SUM, AVG**



```
SELECT COUNT(*) FROM employees;
SELECT AVG(salary) FROM employees;

```







## **SECTION 3: Joins & Relationships**

 **INNER JOIN**



```
SELECT employees.name, departments.department_name  
FROM employees  
INNER JOIN departments ON employees.department_id = departments.id;

```

## **SECTION 4: Grouping & Conditions**



```
SELECT department, COUNT(*)  
FROM employees  
GROUP BY department  
HAVING COUNT(*) > 5;

```

 **CREATE DATABASE / TABLE**
