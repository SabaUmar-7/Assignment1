## 📄 **Project Description**
This project implements a Library Management System (LMS) using object-oriented programming (OOP) principles in Python. It enables the management of books, patrons (students, faculty, etc.), and library operations such as borrowing, returning, and listing books. Key features will include centralized management, automated borrowing/returning, and book search functionalities.

##**Objectives**
**🎯 Centralized Library Management:** Provide a platform to manage books, patrons, and library operations.

**🔄 Automated Borrowing/Returning:** Streamline the process of borrowing and returning books.

**🔍 Book Search:** Enable searching of books based on various criteria like title, author, and genre.

**📚 Patron Management:** Maintain a record of patrons and their borrowed books.

**📖 Simplicity in Handling Books:** Allow adding new books, removing outdated ones, and categorizing books.

## **Scenario Overview**
Imagine you are a librarian at a university's library. You need to manage books, check out books to students and faculty, and ensure that books are returned on time. This system will help you perform these tasks effectively and efficiently.

## **Class: Book**
A Book class will be the template for creating individual books with specific attributes and methods.

### **Example of Attributes :**

*  **Title:** The title of the book (e.g., "Python Programming").
*  **Author:** The author of the book.
*  **Genre:** The genre of the book (e.g., Science, Fiction).
*  **ISBN:** The unique ISBN number.
*  **Availability:** Whether the book is available or currently borrowed.

### **Example of Attributes :**

*  **Check Availability:** A method to check if the book is available.
*  **Borrow Book:** A method to borrow the book (if available).
*  **Return Book:** A method to return the book.


# **▪️ Class Structure:**
## **Base Class: Book**


```python
class Book:
    def __init__(self, title, author, genre, ISBN):
        self.__title = title
        self.__author = author
        self.__genre = genre
        self.__ISBN = ISBN
        self.__availability = True  # Default availability is True

    def get_book_info(self):
        return f"Title: {self.__title}\nAuthor: {self.__author}\nGenre: {self.__genre}\nISBN: {self.__ISBN}"

    def check_availability(self):
        return "Available" if self.__availability else "Not Available"

    def borrow_book(self):
        if self.__availability:
            self.__availability = False
            return f"You have successfully borrowed {self.__title}."
        else:
            return f"{self.__title} is currently not available."

    def return_book(self):
        self.__availability = True
        return f"You have successfully returned {self.__title}."
```

## **Derived Class: AcademicBook**




```python
class AcademicBook(Book):
    def __init__(self, title, author, genre, ISBN, course_name):
        super().__init__(title, author, genre, ISBN)
        self.__course_name = course_name

    def get_course_name(self):
        return f"This book is recommended for the course: {self.__course_name}"

    def get_book_info(self):
        info = super().get_book_info()
        return f"{info}\nCourse Name: {self.__course_name}"
```

## **Derived Class: FictionBook**

```python
class FictionBook(Book):
    def __init__(self, title, author, genre, ISBN, fiction_subgenre):
        super().__init__(title, author, genre, ISBN)
        self.__fiction_subgenre = fiction_subgenre

    def get_fiction_subgenre(self):
        return f"This is a {self.__fiction_subgenre} fiction book."

    def get_book_info(self):
        info = super().get_book_info()
        return f"{info}\nFiction Subgenre: {self.__fiction_subgenre}"
```

## **Derived Class: ReferenceBook**


```python
class ReferenceBook(Book):
    def __init__(self, title, author, genre, ISBN, is_limited_access):
        super().__init__(title, author, genre, ISBN)
        self.__is_limited_access = is_limited_access

    def is_limited_access(self):
        return "Limited access" if self.__is_limited_access else "Unlimited access"

    def get_book_info(self):
        info = super().get_book_info()
        return f"{info}\nAccess: {self.is_limited_access()}"
```

## **Class: Patron**
The Patron class represents a person who borrows books from the library.

```python
class Patron:
    def __init__(self, name, patron_id):
        self.__name = name
        self.__patron_id = patron_id
        self.__borrowed_books = {}

    def get_patron_info(self):
        return f"Patron Name: {self.__name}\nPatron ID: {self.__patron_id}"

    def borrow_book(self, book):
        if book.check_availability() == "Available":
            self.__borrowed_books[book] = book.borrow_book()
            return self.__borrowed_books[book]
        else:
            return f"Sorry, the book '{book.get_book_info()}' is not available."

    def return_book(self, book):
        if book in self.__borrowed_books:
            self.__borrowed_books[book] = book.return_book()
            return self.__borrowed_books[book]
        else:
            return "You haven't borrowed this book."
```

## **Class: Library**

The Library class manages the entire library's collection and patrons.

```python
class Library:
    def __init__(self):
        self.__books = []
        self.__patrons = []

    def add_book(self, book):
        self.__books.append(book)

    def remove_book(self, book):
        if book in self.__books:
            self.__books.remove(book)
        else:
            return "Book not found."

    def add_patron(self, patron):
        self.__patrons.append(patron)

    def list_books(self):
        for book in self.__books:
            print(book.get_book_info())
```

## **Creating Objects (Books and Patrons)**

```python
# Creating book objects
book1 = AcademicBook("Introduction to Python", "John Doe", "Computer Science", "12345", "CS101")
book2 = FictionBook("The Great Gatsby", "F. Scott Fitzgerald", "Fiction", "67890", "Classic")
book3 = ReferenceBook("Encyclopedia of Science", "Various Authors", "Reference", "11223", True)

# Creating patron objects
patron1 = Patron("Alice", "P001")
patron2 = Patron("Bob", "P002")

# Creating a library object
library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Adding patrons to the library
library.add_patron(patron1)
library.add_patron(patron2)

# Borrowing and returning books
print(patron1.borrow_book(book1))  # Alice borrows "Introduction to Python"
print(patron1.return_book(book1))  # Alice returns the book
```
