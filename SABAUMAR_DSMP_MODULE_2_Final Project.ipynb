# ** 1. Import Required Libraries:**





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ** 2. Generate Sample E-commerce Sales Data:**

```python
# Set a random seed for reproducibility
np.random.seed(42)

# Create sample dataset
data = {
    'OrderID': np.arange(1, 101),
    'CustomerID': np.random.randint(1000, 1050, 100),
    'OrderDate': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'ProductName': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch'], 100),
    'Quantity': np.random.randint(1, 5, 100),
    'Price': np.random.randint(100, 1000, 100),
    'PaymentMethod': np.random.choice(['Credit Card', 'Debit Card', 'PayPal', 'Bank Transfer'], 100)
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Create a Total Sales column
df['TotalSales'] = df['Quantity'] * df['Price']

# Display first few rows of dataset
print("Sample Data:")
print(df.head())

# Save dataset as CSV (Optional)
df.to_csv('ecommerce_sales_data.csv', index=False)
```

# ** 3. Revenue Contribution by Customer Regions:**

```python
# Calculate revenue by region
revenue_by_regions = df.groupby('Region')['TotalSales'].sum().reset_index()

# Visualize revenue contribution by top regions
plt.figure(figsize=(9, 4))
sns.barplot(x='Region', y='TotalSales', data=revenue_by_regions, hue='Region', palette="viridis")
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Top Regions for Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# ** 4. Top 5 Best-Selling Products:**

```python
# Calculate total sales by product
top_products = df.groupby('ProductName')['TotalSales'].sum().reset_index().sort_values(by='TotalSales', ascending=False).head(5)

# Visualize top-selling products
plt.figure(figsize=(9, 4))
sns.barplot(x='ProductName', y='TotalSales', data=top_products, palette='coolwarm')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Top 5 Best-Selling Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# ** 5. Monthly Revenue Trend:**

```python
# Extract month from OrderDate
df['Month'] = df['OrderDate'].dt.to_period('M')

# Calculate total sales per month
monthly_revenue = df.groupby('Month')['TotalSales'].sum().reset_index()

# Visualize monthly revenue trend
plt.figure(figsize=(9, 4))
sns.lineplot(x=monthly_revenue['Month'].astype(str), y=monthly_revenue['TotalSales'], marker='o', color='b')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Revenue Trend')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ** 6. Payment Method Distribution:**

```python
# Count payment method usage
payment_counts = df['PaymentMethod'].value_counts()

# Visualize payment method distribution
plt.figure(figsize=(6, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Payment Method Distribution')
plt.show()
```

# ** 7. Customer Purchase Behavior (Top Customers by Revenue):**

```python
# Identify top customers by revenue
top_customers = df.groupby('CustomerID')['TotalSales'].sum().reset_index().sort_values(by='TotalSales', ascending=False).head(5)

# Visualize top customers
plt.figure(figsize=(9, 4))
sns.barplot(x='CustomerID', y='TotalSales', data=top_customers, palette='magma')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales')
plt.title('Top 5 Customers by Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
