## ** Q 1: Basic Class Creation**
### **Task:**

### **Create a Python class called Book that has the following attributes and methods.**

### **(a) Attributes: Title, Author, Pages.**

### **(b) Methods: get_info() that returns a string with the title, author, and number of pages.**

### **Write code to create two objects of the Book class and print their information using the get_info() method.**


````python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def get_info(self):
        return f"'{self.title}' written by {self.author} - {self.pages} pages"


fiction_book = Book('The Great Adventure', 'Mark Twain', 450)
print(fiction_book.get_info())

self_help_book = Book('Mastering Habits', 'Tony Robbins', 550)
print(self_help_book.get_info())
````

**Output :**


```
'The Great Adventure' written by Mark Twain - 450 pages
'Mastering Habits' written by Tony Robbins - 550 pages


```



### **Explanation :**


**1. Class Definition :**

*  The `Book` class contains three attributes: `title`, `author`, and `pages`.

**2. Method :**
* The `get_info()` method constructs a string that includes the title, author, and the number of pages of the book.

**3. Objects :**
Two objects are created:


*  `fiction_book`, which represents a book titled The Great Adventure by Mark Twain with 450 pages.
*  `self_help_book`, which represents a book titled Mastering Habits by Tony Robbins with 550 pages.

## ** Q 2: Class Inheritance**
### **Task:**

### **Create a class Person with the following attributes:**

### **(a) `name, age`**

### **Then create a class Student that inherits from Person and has an additional attribute student_id.**

### **Write a method in Student to display the student's details, including the inherited attributes.**
### **Create an object of Student and print the details.**

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def student_details(self):
        return f"Student Name: {self.name}, Age: {self.age}, ID: {self.student_id}"


student1 = Student('Saba Umar', 19, 67890)
print(student1.student_details())
````

**Output :**


```
Student Name: Saba Umar, Age: 19, ID: 67890

```



### **Explanation :**

**1. Class Definition :**

*  The `Person` class has two attributes: `name` and `age`.

**2. Inheritance :**
*  The `Student` class inherits from the `Person` class and adds a new attribute `student_id`.

**3. Method :**


*  The `student_details()` method in the `Student` class returns the details of the student, including the inherited attributes from `Person` and the new `student_id` attribute.

**4. Object Creation :**


*  An object named `student1` is created with the name Saba Umar, age 19, and student ID 67890.


# ** Q 3: Writing and Reading Files**
## **Task:**
### **Write a Python program to:**

### **(a) Create a text file student_data.txt.**
### **(b) Use a for loop to write the names and scores of 5 students to the file in the format: Name: <"name">, Score: <"score">.**
### **(c) After writing, read the file and print its contents line by line.**

````python
# Writing student data to a file
students = [
    ("Ahmed Shahzad", 85),
    ("Faiza Qureshi", 72),
    ("Usman Tariq", 91),
    ("Sarah Khan", 88),
    ("Zainab Raza", 94)
]

with open('student_data.txt', 'w') as file:
    for name, score in students:
        file.write(f"Name: {name}, Score: {score}\n")

# Reading the student data from the file
with open('student_data.txt', 'r') as file:
    for line in file:
        print(line.strip())

# Example Output
# Name: Ahmed Shahzad, Score: 85
# Name: Faiza Qureshi, Score: 72
# Name: Usman Tariq, Score: 91
# Name: Sarah Khan, Score: 88
# Name: Zainab Raza, Score: 94
````

**Output :**


```
Name: Ahmed Shahzad, Score: 85
Name: Faiza Qureshi, Score: 72
Name: Usman Tariq, Score: 91
Name: Sarah Khan, Score: 88
Name: Zainab Raza, Score: 94



```



### **Explanation :**

**1. File Writing :**

*  The program opens a file named `student_data.txt` in write mode `('w')`. It then uses a `for` loop to take the names and scores of five students as input, writing them to the file in the specified format:` Name: <name>, Score: <score>.`

**2. Example Input :** In this case, the names and scores provided are:
*  Ahmed Shahzad (85)
*  Faiza Qureshi (72)
*  Usman Tariq (91)
*  Sarah Khan (88)
*  Zainab Raza (94)

**3. File Reading :**


*  After writing, the program opens the same file in read mode `('r')` and prints each line from the file, displaying the student names and their corresponding scores.


# ** Q 4: Managing Students Data with OOP & File Handling**
## **Task:**
### **Create a class StudentRecord that handles students records. The class should:**

### **(a) Have methods to add a student's name & score.**
### **(b) Save the record to a text file records.txt.**
### **(c) Have a method to read and print all student records from the file.**
### **Create an instance of StudentRecord and use it to add atleast three students. Then, display all the records from the file.**

````python
class StudentRecord:
    def __init__(self, filename):
        self.filename = filename
        self.students = []

    def save_record(self):
        with open(self.filename, 'w') as file:
            for name, score in self.students:
                file.write(f"{name}: {score}\n")

    def add_student(self, name, score):
        self.students.append((name, score))
        self.save_record()
        print(f"{name} has been added with a score of {score}.")

    def display_record(self):
        print("Student Records:")
        for name, score in self.students:
            print(f"{name}: {score}")

    def read_records(self):
        try:
            with open(self.filename, 'r') as file:
                for line in file:
                    print(line.strip())
        except FileNotFoundError:
            print(f"Error: File '{self.filename}' not found.")
        except Exception as e:
            print(f"An error occurred: {e}")


# Create an instance of StudentRecord and add students
student_records = StudentRecord('records.txt')

# Add students to the records
student_records.add_student('John Doe', 75)
# Output: John Doe has been added with a score of 75.

student_records.add_student('Emma Stone', 88)
# Output: Emma Stone has been added with a score of 88.

student_records.add_student('Liam Smith', 92)
# Output: Liam Smith has been added with a score of 92.

# Read the student records from the file
student_records.read_records()
# Output:
# John Doe: 75
# Emma Stone: 88
# Liam Smith: 92
````

**Output :**


```
John Doe has been added with a score of 75.
Emma Stone has been added with a score of 88.
Liam Smith has been added with a score of 92.
John Doe: 75
Emma Stone: 88
Liam Smith: 92


```


### **Explanation :**

**1. StudentRecord Class :**

*  This class is responsible for managing student records. It has a constructor `(__init__)` that initializes an empty list of students and a file name to store the records.

**2. save_record :**
*  This method saves all the student records to the file. It opens the file in write mode `('w')`, then writes each student's name and score to the file in the format` Name: Score`.

**3. add_student():**


* This method adds a students name and score to the `students` list and then calls the `save_record()` method to save the updated records to the file. It also prints a confirmation message after each addition.


**4. display_record() :**

*   This method simply prints all the student records that are stored in the `students` list.

**5. read_records() :**

*   This method reads and prints the records from the file `(records.txt)`. It opens the file in read mode `('r')` and reads each line, printing it after stripping leading/trailing whitespace. It also handles potential errors, such as when the file is not found.

# ** Q 5: Exception Handling in File Operations**
## **Task:**
### **Write a Python program to:**

### **(a) Attempt to open a file non_existent_file.txt (a file that doesn't exist).**
### **(b) Use Exception Handling to catch the FileNotFoundError and print an error message.**
### **(c) After handling the exception, create the file and a custom message to it.**

````python
try:
    with open('missing_file.txt', 'r') as file:
        print(file.read())
except FileNotFoundError:
    print("Error: The file could not be found.")
    print("Attempting to create a new file...")

    # Creating a new file with a custom message
    with open('new_created_file.txt', 'w') as file:
        file.write("This is a new file created to handle the missing file error. File created successfully.")
        print("New file created and message written successfully.")

# Reading the newly created file
print("Reading the newly created file...")
with open('new_created_file.txt', 'r') as file:
    print(file.read())
````

**Output :**


```
Error: The file could not be found.
Attempting to create a new file...
New file created and message written successfully.
Reading the newly created file...
This is a new file created to handle the missing file error. File created successfully.



```


### **Explanation :**

**1. Try-Except Block :**

*  The program tries to open a file that does not exist `(missing_file.txt)`. If the file is not found, a `FileNotFoundError` exception is raised, and the program catches this error.

**2. Error Handling :**
*  The `except` block catches the `FileNotFoundError` and prints a custom error message. After that, the program proceeds to create a new file named `new_created_file.txt` and writes a custom message into it, confirming the successful creation of the file..

**3. File Creation :**


* The program opens the `new_created_file.txt` in write mode `('w')`, writes the message `"This is a new file created to handle the missing file error. File created successfully."`, and then confirms that the file was created successfully.


**4. Reading the Created File :**

*   The program then opens the newly created file in read mode `('r')`, reads its contents, and prints them to the console.
