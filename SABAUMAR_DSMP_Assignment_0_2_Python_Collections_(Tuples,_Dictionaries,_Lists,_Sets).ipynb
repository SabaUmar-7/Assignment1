## ** Q 1: Create and Count Attendance List**
### **Task:**

### **(a) Create a list attendance_list with names, some repeated.**

### **(b) Print the list.**

## ** Q 2: Unique Attendance using a Set**
### **Task:**

### **(a) Convert the attendance_list into a set to find unique attendees (sets remove duplicates).**

### **(b) Print the set to display unique student names.**

### **(c) Count and print the total number of unique attendees.**

````python
# Step 1: Creating a list of student names for attendance check
attendance_list = ["Michael", "Sarah", "Michael", "Olivia",
    "Liam", "Sarah", "Saba", "Noah",
    "Ava", "Liam", "Sophia", "James"]

# Step 2: Converting the attendance list to a set to find unique attendees
unique_attendees = set(attendance_list)

# Step 3: Printing the set to display unique student names
print("Unique Attendees:", unique_attendees)

# Step 4: Counting and printing the total number of unique attendees
total_unique_attendees = len(unique_attendees)
print("Total number of unique attendees:", total_unique_attendees)
````

**Output :**


```
Unique Attendees: {'Noah', 'Liam', 'Olivia', 'Ava', 'James', 'Sophia', 'Saba', 'Michael', 'Sarah'}
Total number of unique attendees: 9
```



### **Explanation :**

**1. Creating the Attendance List :**

*  An `attendance_list` is defined, containing names with duplicates.

**2. Finding Unique Attendees :**
*  The list is converted into a set named `unique_attendees`, which automatically removes any duplicate names.

**3. Printing Unique Names :**


*  The ` print()` function displays the set of unique student names.

**4. Counting Unique Attendees :**


*  The `len()` function counts the number of unique attendees in the set and prints the total.

# ** Q 4: Count Attendance using a Dictionary**
## **Task:**
### **(a) Using the attendance_list, create a dictionary where each student's name is a key, and the value is the count of how many times they attended.**

### **(b) Print the dictionary to see each student's attendance count. Solution.**

````python
# Step 1: Using a modified attendance_list
attendance_list = ["Michael", "Sarah", "Michael", "Olivia", "Liam", "Sarah", "Liam", "Sophia"]

# Step 2: Creating an empty dictionary to store the attendance count
attendance_count = {}

# Step 3: Counting the number of times each student attended
for student in attendance_list:
    if student in attendance_count:
        attendance_count[student] += 1
    else:
        attendance_count[student] = 1

# Step 4: Printing the dictionary to see each student's attendance count
print("Attendance Count Dictionary:", attendance_count)
````

**Output :**


```
Attendance Count Dictionary: {'Michael': 2, 'Sarah': 2, 'Olivia': 1, 'Liam': 2, 'Sophia': 1}


```


### **Explanation :**

**1. Using a Dictionary  :**

*  A dictionary is ideal because it stores key-value pairs. The student names act as keys, and the attendance counts are the corresponding values.

**2. Iterating Through the List :**
*  The` for` loop goes through each name in the `attendance_list`.

**3. Updating Attendance Counts:**


*  If a students name already exists as a key in the attendance_count dictionary, their value `(attendance count)` is incremented by 1.
*    If the name is not found, a new key-value pair is added with the initial count of 1.


**4. Final Dictionary :**

*  The result is a dictionary showing the number of times each student attended. This is printed using the `print()` function.
