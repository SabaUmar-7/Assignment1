## ▪️ **Exercise 6: Built-in Functions**
### **Write a program that:**

* ### **Takes a list of numbers as input from the user.**

* ### **Uses the `abs()`, `max()`, and `min()` functions to find the absolute values, maximum, and minimum of the numbers.**

* ### **Prints the results with appropriate messages.**


````python
# Take a list of numbers as input from the user
user_input = input("Enter a list of numbers separated by spaces: ")

# Convert the input string to a list of numbers (float)
number_list = [float(num) for num in user_input.split()]

# Find absolute values, maximum, and minimum
absolute_values = [abs(num) for num in number_list]
maximum_value = max(number_list)
minimum_value = min(number_list)

# Print the results with appropriate messages
print("Absolute values of the numbers:", absolute_values)
print("Maximum value:", maximum_value)
print("Minimum value:", minimum_value)
````

### **Explanation :**

* **User Input:** The program prompts the user to enter a list of numbers separated by spaces.

* **Convert Input to List of Numbers:** The input string is split into individual components and converted to a list of floats.

* **Calculate Absolute Values:** It calculates the absolute values of each number using the `abs()` function.

* **Find Maximum and Minimum:** The program uses `max()` and `min()` functions to find the maximum and minimum values in the list.

* **Print Results:** Finally, the program prints the absolute values, maximum, and minimum values with appropriate messages.

**Example :**

If the user enters `-5 3 -1 7.5 0`, the output will be:



```
Absolute values of the numbers: [5.0, 3.0, 1.0, 7.5, 0.0]
Maximum value: 7.5
Minimum value: -5.0

```



## **▪️ Exercise 7: Arithmetic Operations**

### **Create a simple calculator program that:**

* ### **Prompts the user to enter two numbers.**

* ### **Asks the user to select an operation: addition, subtraction, multiplication, or division.**

* ### **Performs the selected operation and prints the result. Handle division by zero gracefully.**


````python
# Simple Calculator Program

# Function to perform arithmetic operations
def calculator():
    # Prompt the user to enter two numbers
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input! Please enter numeric values.")
        return

    # Ask the user to select an operation
    print("Select an operation:")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")

    operation = input("Enter the number corresponding to the operation: ")

    # Perform the selected operation and print the result
    if operation == '1':
        result = num1 + num2
        print("Result: ", result)
    elif operation == '2':
        result = num1 - num2
        print("Result: ", result)
    elif operation == '3':
        result = num1 * num2
        print("Result: ", result)
    elif operation == '4':
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
        else:
            result = num1 / num2
            print("Result: ", result)
    else:
        print("Invalid operation selected!")

# Run the calculator
calculator()
````

### **Explanation :**

* **Input Two Numbers:**
The program prompts us to input two numbers, converting them to floats. If we enter non-numeric values, a `ValueError` is raised, and an error message is displayed.

* **Select an Operation:**
We can choose from four arithmetic operations: addition, subtraction, multiplication, and division by entering the corresponding number.

* **Perform the Selected Operation:**
The program checks our selection and performs the corresponding operation, displaying the result. If an invalid operation is selected (e.g., entering `5`), the output will be:

```
   Invalid operation selected!

```
For division, it checks if the second number is zero to avoid errors.

*  **Output:** The result of the operation or an error message is displayed based on our input.

**Example :**

Entering 10, 5, and selecting division results in :


```
Result:  2.0

```
If we attempts to divide by zero, the output will be:



```
Error: Division by zero is not allowed.

```
